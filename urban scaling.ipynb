```python
def info(data, city):
    value = []
    for i in range(14):
        meme = data[i+2006][city]
        value.append(meme)
    return value
```

```python
def popinfo(data, city):
    value = []
    for i in range(14):
        meme = data[str(i+2006)][city]
        value.append(meme)
    return value
```
